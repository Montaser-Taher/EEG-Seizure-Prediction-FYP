```python
from pathlib import Path
import pandas as pd
import sys

sys.path.append(str(Path("../src").resolve()))

from config import (
    RAW_ROOT,
    GROUP4_PATIENTS,
    BAD_FILES,
    CHANNEL_ORDER
)

from utils import (
    load_raw_edf,
    prepare_raw,
    reorder_channels
)

# same exclusion as your pipeline
EXCLUDED_GROUP4_PATIENTS = ["chb12"]

GROUP4_SELECTED_PATIENTS = [
    patient for patient in GROUP4_PATIENTS
    if patient not in EXCLUDED_GROUP4_PATIENTS
]

rows = []
patient_channel_summary = []

for patient_name in GROUP4_SELECTED_PATIENTS:
    print("\n" + "=" * 80)
    print(f"Checking patient: {patient_name}")

    patient_folder = RAW_ROOT / patient_name
    edf_files = sorted(patient_folder.glob("*.edf"))
    patient_bad_files = set(BAD_FILES.get(patient_name, []))

    # keep only valid files
    patient_valid_files = []
    for file_path in edf_files:
        if file_path.name in patient_bad_files:
            continue
        patient_valid_files.append(file_path)

    if len(patient_valid_files) == 0:
        print(f"No valid files for {patient_name}")
        continue

    # build patient-specific common channel set exactly like your pipeline
    first_raw = load_raw_edf(patient_valid_files[0])
    first_raw = prepare_raw(first_raw)
    reference_channels = first_raw.ch_names.copy()

    patient_common_channels = set(reference_channels)

    for file_path in patient_valid_files[1:]:
        raw_tmp = load_raw_edf(file_path)
        raw_tmp = prepare_raw(raw_tmp)
        current_channels = set(raw_tmp.ch_names)
        patient_common_channels = patient_common_channels.intersection(current_channels)

    # keep order close to standard order
    expected_channels = [ch for ch in CHANNEL_ORDER if ch in patient_common_channels]

    # add extra common channels not already in CHANNEL_ORDER
    extra_common_channels = [
        ch for ch in reference_channels
        if ch in patient_common_channels and ch not in expected_channels
    ]
    expected_channels.extend(extra_common_channels)

    print(f"Patient-specific final common channels ({len(expected_channels)}):")
    print(expected_channels)

    patient_channel_summary.append({
        "patient_name": patient_name,
        "n_patient_common_channels": len(expected_channels),
        "patient_common_channels": ", ".join(expected_channels)
    })

    # now check every valid file after reorder
    for file_path in patient_valid_files:
        raw = load_raw_edf(file_path)
        raw = prepare_raw(raw)
        raw = reorder_channels(raw, expected_channels)

        rows.append({
            "patient_name": patient_name,
            "file_name": file_path.name,
            "n_channels_after_reorder": len(raw.ch_names),
            "channels_after_reorder": ", ".join(raw.ch_names)
        })

# build dataframes
df_files = pd.DataFrame(rows)
df_patients = pd.DataFrame(patient_channel_summary)

# save results
df_files.to_csv("group4_file_channel_check_after_processing_logic.csv", index=False)
df_patients.to_csv("group4_patient_common_channel_sets.csv", index=False)

print("\nSaved:")
print("- group4_file_channel_check_after_processing_logic.csv")
print("- group4_patient_common_channel_sets.csv")

print("\nPatient-level summary:")
print(df_patients)

print("\nFirst few file-level rows:")
print(df_files.head())
```

```python
from pathlib import Path
import pandas as pd
import sys

sys.path.append(str(Path("../src").resolve()))

from config import RAW_ROOT, GROUP4_PATIENTS, BAD_FILES, CHANNEL_ORDER
from utils import load_raw_edf, prepare_raw

EXCLUDED_GROUP4_PATIENTS = ["chb12"]

GROUP4_SELECTED_PATIENTS = [
    patient for patient in GROUP4_PATIENTS
    if patient not in EXCLUDED_GROUP4_PATIENTS
]

print("Selected Group 4 patients:")
print(GROUP4_SELECTED_PATIENTS)

group4_final_channel_rows = []
group4_final_channel_sets = {}

for patient_name in GROUP4_SELECTED_PATIENTS:
    patient_folder = RAW_ROOT / patient_name
    edf_files = sorted(patient_folder.glob("*.edf"))
    patient_bad_files = set(BAD_FILES.get(patient_name, []))

    valid_files = [f for f in edf_files if f.name not in patient_bad_files]

    print("\n" + "=" * 80)
    print(f"Patient: {patient_name}")
    print(f"Valid files used: {len(valid_files)}")

    if len(valid_files) == 0:
        print("No valid files found.")
        group4_final_channel_sets[patient_name] = []
        continue

    patient_common = None
    reference_channels = None

    for i, file_path in enumerate(valid_files):
        raw = load_raw_edf(file_path)
        raw = prepare_raw(raw)
        current_channels = set(raw.ch_names)

        if i == 0:
            reference_channels = raw.ch_names.copy()
            patient_common = current_channels
        else:
            patient_common = patient_common.intersection(current_channels)

    # keep same ordering logic as preprocessing
    expected_channels = [ch for ch in CHANNEL_ORDER if ch in patient_common]

    extra_common_channels = [
        ch for ch in reference_channels
        if ch in patient_common and ch not in expected_channels
    ]
    expected_channels.extend(extra_common_channels)

    group4_final_channel_sets[patient_name] = expected_channels

    print(f"Final channel count: {len(expected_channels)}")
    print(expected_channels)

    group4_final_channel_rows.append({
        "patient_name": patient_name,
        "n_final_channels": len(expected_channels),
        "final_channels": ", ".join(expected_channels)
    })

df_group4_final_channels = pd.DataFrame(group4_final_channel_rows)
display(df_group4_final_channels)

for patient_name, ch_list in group4_final_channel_sets.items():
    print("\n" + "=" * 80)
    print(f"Patient: {patient_name}")
    print(f"Number of final channels: {len(ch_list)}")
    print(ch_list)
```

```python
group123_shared_17 = [
    "FP1-F7", "F7-T7", "T7-P7", "P7-O1",
    "FP1-F3", "F3-C3", "C3-P3", "P3-O1",
    "FP2-F4", "F4-C4", "C4-P4", "P4-O2",
    "FP2-F8", "F8-T8", "P8-O2", "FZ-CZ", "CZ-PZ"
]

group123_shared_17_set = set(group123_shared_17)

rows = []

for patient_name, ch_list in group4_final_channel_sets.items():
    ch_set = set(ch_list)
    overlap = [ch for ch in group123_shared_17 if ch in ch_set]
    missing = [ch for ch in group123_shared_17 if ch not in ch_set]

    rows.append({
        "patient_name": patient_name,
        "n_final_channels": len(ch_list),
        "n_overlap_with_main_17": len(overlap),
        "can_match_full_17": len(overlap) == 17,
        "missing_from_main_17": ", ".join(missing)
    })

df_group4_final_vs_17 = pd.DataFrame(rows)
display(df_group4_final_vs_17)

#so after seeing the final channel set could over lap with the 7 channel set as the other groups
```
