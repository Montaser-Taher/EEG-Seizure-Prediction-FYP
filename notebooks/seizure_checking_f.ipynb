```python
from pathlib import Path
import pandas as pd
import sys

sys.path.append(str(Path("../src").resolve()))

from config import (
    RAW_ROOT,
    GROUP1_PATIENTS,
    GROUP2_PATIENTS,
    GROUP3_PATIENTS,
    GROUP4_PATIENTS
)

from utils import parse_chbmit_summary

all_patients = (
    GROUP1_PATIENTS +
    GROUP2_PATIENTS +
    GROUP3_PATIENTS +
    GROUP4_PATIENTS
)

seizure_rows = []

for patient_name in all_patients:
    patient_folder = RAW_ROOT / patient_name
    summary_path = patient_folder / f"{patient_name}-summary.txt"

    if not summary_path.exists():
        seizure_rows.append({
            "patient_name": patient_name,
            "n_files_with_seizures": 0,
            "n_total_seizure_intervals": 0,
            "has_seizure": False
        })
        continue

    summary_dict = parse_chbmit_summary(summary_path)

    n_files_with_seizures = 0
    n_total_seizure_intervals = 0

    for file_name, seizure_intervals in summary_dict.items():
        if len(seizure_intervals) > 0:
            n_files_with_seizures += 1
            n_total_seizure_intervals += len(seizure_intervals)

    seizure_rows.append({
        "patient_name": patient_name,
        "n_files_with_seizures": n_files_with_seizures,
        "n_total_seizure_intervals": n_total_seizure_intervals,
        "has_seizure": n_total_seizure_intervals > 0
    })

df_patient_seizure_check = pd.DataFrame(seizure_rows)
display(df_patient_seizure_check.sort_values(["has_seizure", "n_total_seizure_intervals"], ascending=[False, False]))

print("Patients WITH seizures:")
display(df_patient_seizure_check[df_patient_seizure_check["has_seizure"] == True])

print("\nPatients WITHOUT seizures:")
display(df_patient_seizure_check[df_patient_seizure_check["has_seizure"] == False])
```
