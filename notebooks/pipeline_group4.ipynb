```python
from pathlib import Path
import pandas as pd
import sys

sys.path.append(str(Path("../src").resolve()))

from config import RAW_ROOT, GROUP4_PATIENTS, BAD_FILES, CHANNEL_ORDER
from utils import load_raw_edf, prepare_raw

print("Checking Group 4 patients...")
print("Group 4 patients:", GROUP4_PATIENTS)

print("Checking Group 4 patients...")
print("Group 4 patients:", GROUP4_PATIENTS)

group4_file_rows = []

for patient_name in GROUP4_PATIENTS:
    patient_folder = RAW_ROOT / patient_name
    edf_files = sorted(patient_folder.glob("*.edf"))
    patient_bad_files = BAD_FILES.get(patient_name, [])

    print("\n" + "=" * 80)
    print(f"Patient: {patient_name}")
    print(f"Total EDF files found: {len(edf_files)}")
    print(f"Known bad files from config: {patient_bad_files}")

    for file_path in edf_files:
        file_name = file_path.name

        raw = load_raw_edf(file_path)
        raw = prepare_raw(raw)

        group4_file_rows.append({
            "patient_name": patient_name,
            "file_name": file_name,
            "n_eeg_channels": len(raw.ch_names),
            "channels": ", ".join(raw.ch_names),
            "is_known_bad": file_name in patient_bad_files
        })

df_group4_files = pd.DataFrame(group4_file_rows)
display(df_group4_files[["patient_name", "file_name", "n_eeg_channels", "is_known_bad"]])
```

```python
group4_patient_common = {}
group4_patient_summary_rows = []

for patient_name in GROUP4_PATIENTS:
    patient_folder = RAW_ROOT / patient_name
    edf_files = sorted(patient_folder.glob("*.edf"))
    patient_bad_files = set(BAD_FILES.get(patient_name, []))

    valid_files = [f for f in edf_files if f.name not in patient_bad_files]

    print("\n" + "=" * 80)
    print(f"Patient: {patient_name}")
    print(f"Valid files used for common-channel check: {len(valid_files)}")

    if len(valid_files) == 0:
        print("No valid files found after skipping known bad files.")
        group4_patient_common[patient_name] = set()
        continue

    patient_common = None

    for file_path in valid_files:
        raw = load_raw_edf(file_path)
        raw = prepare_raw(raw)
        current_channels = set(raw.ch_names)

        if patient_common is None:
            patient_common = current_channels
        else:
            patient_common = patient_common.intersection(current_channels)

    group4_patient_common[patient_name] = patient_common

    print(f"Common channels inside {patient_name}: {len(patient_common)}")
    print(sorted(patient_common))

    group4_patient_summary_rows.append({
        "patient_name": patient_name,
        "n_valid_files": len(valid_files),
        "n_common_channels_within_patient": len(patient_common),
        "common_channels_within_patient": ", ".join(sorted(patient_common))
    })

df_group4_patient_summary = pd.DataFrame(group4_patient_summary_rows)
display(df_group4_patient_summary)
```

```python
for patient_name, ch_set in group4_patient_common.items():
    print("\n" + "=" * 80)
    print(f"Patient: {patient_name}")
    print(f"Number of common channels: {len(ch_set)}")
    print(sorted(ch_set))
```

```python
group4_global_common = None

for patient_name, ch_set in group4_patient_common.items():
    if len(ch_set) == 0:
        continue

    if group4_global_common is None:
        group4_global_common = set(ch_set)
    else:
        group4_global_common = group4_global_common.intersection(ch_set)

if group4_global_common is None:
    group4_global_common = set()

print("\n" + "=" * 80)
print("Possible global common channels across ALL Group 4 patients:")
print(f"Count: {len(group4_global_common)}")
print(sorted(group4_global_common))
```
